```python
# Matrix Multiplication (using loops and numpy)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def isValid(A, B):
    S1 = A.shape
    S2 = B.shape
    if S1[1] == S2[0]:
        return True
    else:
        return False


def matrix_multiply(A, B):
    m, n = A.shape
    o = B.shape[1]
    ans = np.zeros((m, o))

    for i in range(m):
        for j in range(o):
            for k in range(n):
                ans[i][j] += A[i][k] * B[k][j]

    return ans


def matrix_multiply_2(A, B):
    return A @ B

A = np.array([[2, 4], [6, 8]])

B = np.array([[1, 2], [3, 4]])

if isValid(A, B):
    print(f"Result using loops: {matrix_multiply(A,B)}")
    print(f"Result using numpy: {matrix_multiply_2(A,B)}")
else:
    print(f"Matrix multiplication is not valid")
```

```python
# Sigmoid and its derivative
import numpy as np


def sigmoidfn(x):
    arr = []
    for i in x:
        arr.append(1 / (1 + np.exp(-i)))

    return arr


def derivative(x):
    arr = []
    for i in x:
        arr.append(np.exp(-i) / ((1 + np.exp(-i)) ** 2))
    return arr


x = np.array([[9, 2, 5, 0, 0], [7, 5, 0, 0, 0]])
print(f"x on applying sigmoid activation fn is: {sigmoidfn(x)}")
print(f"x on applying derivative of sigmoid activation fn is: {derivative(x)}")
```

```python
# Z-Score Normalisation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def mean(x):
    return sum(x) / x.size


def standard_deviation(x):
    var = 0
    av = mean(x)
    for i in x:
        var += (i - av) ** 2
    return (var / x.size) ** 0.5


def zscore_normalisation(x):
    av = mean(x)
    sd = standard_deviation(x)
    arr = np.empty(x.size)
    for i in range(x.size):
        arr[i] = (x[i] - av) / sd
    return arr


def zscore_normalisation_2(x):
    av = np.mean(x)
    sd = np.std(x)
    arr = np.empty(x.size)
    for i in range(x.size):
        arr[i] = (x[i] - av) / sd
    return arr


def zscore_normalisation_3(x):
    ans = (x - np.mean(x)) / np.std(x)

    return ans


x = np.array([1, 2, 3])
print(mean(x))
print(standard_deviation(x))
print(zscore_normalisation(x))
print(zscore_normalisation_2(x))
print(zscore_normalisation_3(x))
```
